```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

## Exercice 1: Utilisation de scikit-learn pour la regression lineaire


```python
#generation de donnees test
n = 100
x = np.arange(n) # x is in (5,) shape
y = np.random.randn(n)*30 + 50. * np.log(1 + np.arange(n))

# instanciation de sklearn.linear_model.LinearRegression
lr = LinearRegression()
lr.fit(x[:, np.newaxis], y)  # np.newaxis est utilise car x doit etre une matrice 2d avec 'LinearRegression'
# x[:, np.newaxis] transforms x from shape (5,) to shape (5,1)

# representation du resultat
fig = plt.figure()
plt.plot(x, y, 'r.')
plt.plot(x, lr.predict(x[:, np.newaxis]), 'b-')
plt.legend(('Data', 'Linear Fit'), loc='lower right')
plt.title('Linear regression')
plt.show()
```

### <span style="color:blue">QUESTION 1.1 :</span> 
Expliquer ce qu'est *lr* et ce que font *lr.fit* et *lr.predict*



### <span style="color:blue">REPONSE 1.1 :</span> 
*lr* est l'objet sikit-learn qui nous permet de créer de régressions linéaires. Alors, le *lr.fit* va fitter notre modèle aux données, c'est à dire qu'il va modifier les paramètres $\Theta$ de notre modèle, dans ce cas, les coefficients linéaire et angulaire de la régression de tel forme qu'ils minimisent les moindres carrés ordinaires par rapport à nos données.

En conséquence, le *lr.predict* va faire une prévision de sortie étant donnée une entrée, dans ce cas, il va nous donner la valeur $y$ du point sur la ligne qui a $x=x_{donnée}$.
